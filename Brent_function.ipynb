```
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
import random
from mpl_toolkits.mplot3d import Axes3D 
```

```
def funcao_objetivo(x, y):
  return 0.26 * (x**2 + y**2) - 0.48*x*y
```

Gera o gráfico da Função Objetivo

```
x = []
y = []
a = []
b = []

for i in range(-10,10):
  y.append(funcao_objetivo(a, b))
  x.append(i)
  
# Gerar o gráfico de linhas
plt.plot(x, y )

# Título
plt.title("Quadratic Function")

plt.ylabel("Value Function")
plt.xlabel("X")

# Salva imagem
plt.savefig('line_graph')

plt.close('all')

# Exibe imagem
Image('line_graph.png')
```

# Algoritmo de busca aleatória

```
# Nova solução
interacoes = 100

# Gera o gráfico - Lista
func_objetivo = []
tentativas = []
valor_x = []
valor_y = []

i = 1

x = random.uniform(- 10, 10)
y = random.uniform(- 10, 10)

fo = funcao_objetivo(x, y)

# Salva os dados
func_objetivo.append(fo)
valor_x.append(x)
valor_y.append(y)
tentativas.append(0)

while i < interacoes:
  
  # Gera uma nova solucao de forma aleatória
  temp_x = random.uniform(- 10, 10)
  temp_y = random.uniform(- 10, 10)
  
  temp_fo = funcao_objetivo(temp_x, temp_y)
  
  # Se a nova solucao for Melhor
  if( temp_fo < fo ):
    fo = temp_fo
    x = temp_x
    y = temp_y
    
  # Salva os dados
  func_objetivo.append(fo)
  tentativas.append(i)
  valor_x.append(x)
  valor_y.append(y)
  
  #Tentativa
  i = i + 1
print(func_objetivo)

y = func_objetivo
x = tentativas

# Gera o gráfico de linhas
plt.plot(x, y )

# Título
plt.title("Random_Search Performance")

plt.ylabel("Value Objetive_Function")
plt.xlabel("Interactions")

# Salva imagem
plt.savefig('line_graph')

plt.close('all')

# Exibe imagem
Image('line_graph.png')
```

Método de Gradiente


```
interacoes = 100

func_objetivo = []
tentativas = []
valor_x = []
valor_y = []

i = 1
 
x = random.uniform(- 10, 10)
y = random.uniform(- 10, 10)

# Qualidade da solução
fo = funcao_objetivo(x, y)

func_objetivo.append(fo)
valor_x.append(x)
valor_y.append(y)
tentativas.append(0)

while i < interacoes:
  
  # Gera nova solução
  temp_x = x + random.uniform(- 0.5, 0.5) 
  temp_y = y + random.uniform(- 0.5, 0.5)
  
  temp_fo = funcao_objetivo(temp_x, temp_y)
  
  if( temp_fo < fo ):
    fo = temp_fo
    x = temp_x
    y = temp_y
  
  func_objetivo.append(fo)
  tentativas.append(i)
  valor_x.append(x)
  
  i = i + 1

y = func_objetivo
x = tentativas

# Gera o gráfico de linhas
plt.plot(x, y )

plt.title("Search_Gradient_Method Performance")

plt.ylabel("Value Objetive_Function")
plt.xlabel("Interactions")

plt.savefig('line_graph')

plt.close('all')

Image('line_graph.png')
```

```
def fun(x, y):
    z = []
    for i in range(len(x)):
      z.append(funcao_objetivo(x[i], y[i]))
    return z

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-10.0, 10.0, 0.05)
X, Y = np.meshgrid(x, y)
print(fun(np.ravel(X), np.ravel(Y)))
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Salva imagem
plt.savefig('line_graph')

plt.close('all')

# Exibe imagem
Image('line_graph.png')
```
